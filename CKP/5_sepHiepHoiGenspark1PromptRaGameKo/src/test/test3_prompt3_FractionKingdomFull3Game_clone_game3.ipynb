```bash
<!DOCTYPE html>
<html lang="en">
<head>
    <meta charset="UTF-8">
    <meta name="viewport" content="width=device-width, initial-scale=1.0">
    <title>Fraction Learning: Circle Slicing</title>
    <style>
        body {
            font-family: 'Comic Sans MS', cursive, sans-serif;
            background-color: #f0f8ff;
            margin: 0;
            padding: 20px;
            display: flex;
            flex-direction: column;
            align-items: center;
            overflow: hidden;
        }
        
        h1 {
            color: #4b0082;
            text-align: center;
            margin-bottom: 10px;
            font-size: 28px;
            animation: bounce 1s ease infinite alternate;
        }
        
        @keyframes bounce {
            from { transform: translateY(0px); }
            to { transform: translateY(-5px); }
        }
        
        .instructions {
            background-color: #fff;
            border: 3px solid #9370db;
            border-radius: 15px;
            padding: 15px;
            margin-bottom: 20px;
            max-width: 780px;
            text-align: center;
            font-size: 18px;
            color: #333;
            box-shadow: 0 5px 15px rgba(0,0,0,0.1);
            transition: all 0.3s ease;
        }
        
        .instructions:hover {
            transform: translateY(-5px);
            box-shadow: 0 10px 20px rgba(0,0,0,0.15);
        }
        
        .progress-bar {
            display: flex;
            justify-content: space-between;
            width: 780px;
            margin-bottom: 20px;
            background-color: #fff;
            border: 3px solid #9370db;
            border-radius: 15px;
            padding: 10px 15px;
            box-shadow: 0 5px 15px rgba(0,0,0,0.1);
        }
        
        .progress-step {
            display: flex;
            flex-direction: column;
            align-items: center;
            cursor: pointer;
            transition: all 0.3s ease;
        }
        
        .progress-step:hover {
            transform: translateY(-5px);
        }
        
        .step-circle {
            width: 50px;
            height: 50px;
            border-radius: 50%;
            background-color: #e0e0e0;
            display: flex;
            justify-content: center;
            align-items: center;
            font-size: 20px;
            font-weight: bold;
            color: #333;
            margin-bottom: 5px;
            border: 3px solid #9370db;
            transition: all 0.3s ease;
        }
        
        .step-circle.active {
            background-color: #9370db;
            color: white;
        }
        
        .step-circle.completed {
            background-color: #32cd32;
            color: white;
        }
        
        .step-label {
            font-size: 14px;
            color: #333;
            text-align: center;
        }
        
        .game-container {
            position: relative;
            width: 800px;
            height: 800px;
            background-color: white;
            border: 5px solid #9370db;
            border-radius: 15px;
            overflow: hidden;
            box-shadow: 0 10px 20px rgba(0,0,0,0.2);
            transition: all 0.3s ease;
        }
        
        .game-container:hover {
            box-shadow: 0 15px 30px rgba(0,0,0,0.3);
        }
        
        .canvas-container {
            position: relative;
            width: 100%;
            height: 100%;
        }
        
        canvas {
            position: absolute;
            top: 0;
            left: 0;
        }
        
        .challenge-container {
            position: absolute;
            top: 20px;
            left: 20px;
            background-color: rgba(255, 255, 255, 0.9);
            border: 3px solid #ff69b4;
            border-radius: 10px;
            padding: 10px;
            font-size: 20px;
            color: #4b0082;
            z-index: 10;
            box-shadow: 0 5px 10px rgba(0,0,0,0.1);
            transition: all 0.3s ease;
            animation: pulse 2s infinite;
        }
        
        @keyframes pulse {
            0% { transform: scale(1); }
            50% { transform: scale(1.05); }
            100% { transform: scale(1); }
        }
        
        .feedback-container {
            position: absolute;
            top: 50%;
            left: 50%;
            transform: translate(-50%, -50%) scale(0.9);
            background-color: rgba(255, 255, 255, 0.95);
            border: 3px solid #32cd32;
            border-radius: 15px;
            padding: 20px;
            font-size: 28px;
            color: #32cd32;
            text-align: center;
            z-index: 20;
            display: none;
            box-shadow: 0 10px 30px rgba(0,0,0,0.2);
            animation: pop-in 0.5s forwards;
        }
        
        @keyframes pop-in {
            0% { transform: translate(-50%, -50%) scale(0.5); opacity: 0; }
            70% { transform: translate(-50%, -50%) scale(1.1); opacity: 1; }
            100% { transform: translate(-50%, -50%) scale(1); opacity: 1; }
        }
        
        .progress-container {
            position: absolute;
            bottom: 20px;
            right: 20px;
            background-color: rgba(255, 255, 255, 0.9);
            border: 3px solid #4b0082;
            border-radius: 10px;
            padding: 10px;
            font-size: 18px;
            color: #4b0082;
            z-index: 10;
            box-shadow: 0 5px 10px rgba(0,0,0,0.1);
        }
        
        .button {
            background-color: #ff69b4;
            color: white;
            border: none;
            border-radius: 10px;
            padding: 10px 20px;
            font-size: 18px;
            cursor: pointer;
            z-index: 10;
            font-family: 'Comic Sans MS', cursive, sans-serif;
            box-shadow: 0 5px 10px rgba(0,0,0,0.1);
            transition: all 0.2s ease;
            position: absolute;
        }
        
        .button:hover {
            transform: translateY(-3px);
            box-shadow: 0 8px 15px rgba(0,0,0,0.2);
        }
        
        .button:active {
            transform: translateY(1px);
            box-shadow: 0 2px 5px rgba(0,0,0,0.1);
        }
        
        .reset-button {
            bottom: 20px;
            left: 20px;
            background-color: #ff69b4;
        }
        
        .next-button {
            bottom: 20px;
            left: 150px;
            background-color: #32cd32;
            display: none;
        }
        
        .hint-button {
            top: 20px;
            right: 20px;
            background-color: #4b0082;
        }
        
        .confetti {
            position: absolute;
            width: 10px;
            height: 10px;
            background-color: #f00;
            border-radius: 50%;
            pointer-events: none;
            z-index: 30;
        }
        
        .fraction-visual {
            position: absolute;
            top: 70px;
            left: 20px;
            background-color: rgba(255, 255, 255, 0.95);
            border: 3px solid #ff69b4;
            border-radius: 10px;
            padding: 10px;
            font-size: 18px;
            color: #4b0082;
            z-index: 10;
            display: none;
            box-shadow: 0 8px 15px rgba(0,0,0,0.15);
            animation: slide-in 0.5s forwards;
        }
        
        @keyframes slide-in {
            from { transform: translateX(-50px); opacity: 0; }
            to { transform: translateX(0); opacity: 1; }
        }
        
        .fraction-visual svg {
            width: 100px;
            height: 100px;
            animation: rotate 1s ease-out;
        }
        
        @keyframes rotate {
            from { transform: rotate(-30deg); opacity: 0; }
            to { transform: rotate(0); opacity: 1; }
        }
        
        .temp-message {
            position: absolute;
            top: 50%;
            left: 50%;
            transform: translate(-50%, -50%);
            background-color: rgba(255, 255, 255, 0.95);
            border: 3px solid #ff6347;
            border-radius: 15px;
            padding: 20px;
            font-size: 24px;
            color: #ff6347;
            text-align: center;
            z-index: 20;
            box-shadow: 0 10px 20px rgba(0,0,0,0.2);
            animation: fade-in-out 1.5s forwards;
        }
        
        @keyframes fade-in-out {
            0% { opacity: 0; transform: translate(-50%, -50%) scale(0.8); }
            20% { opacity: 1; transform: translate(-50%, -50%) scale(1.1); }
            80% { opacity: 1; transform: translate(-50%, -50%) scale(1); }
            100% { opacity: 0; transform: translate(-50%, -50%) scale(0.8); }
        }
        
        .completion-message {
            position: absolute;
            top: 50%;
            left: 50%;
            transform: translate(-50%, -50%);
            background-color: rgba(255, 255, 255, 0.95);
            border: 3px solid #32cd32;
            border-radius: 15px;
            padding: 30px;
            font-size: 32px;
            color: #32cd32;
            text-align: center;
            z-index: 30;
            box-shadow: 0 15px 30px rgba(0,0,0,0.3);
            animation: scale-in 0.7s forwards;
        }
        
        @keyframes scale-in {
            0% { transform: translate(-50%, -50%) scale(0.1); opacity: 0; }
            60% { transform: translate(-50%, -50%) scale(1.1); opacity: 1; }
            100% { transform: translate(-50%, -50%) scale(1); opacity: 1; }
        }
        
        .restart-button {
            background-color: #32cd32;
            color: white;
            border: none;
            border-radius: 10px;
            padding: 10px 20px;
            font-size: 20px;
            margin-top: 20px;
            cursor: pointer;
            font-family: 'Comic Sans MS', cursive, sans-serif;
            box-shadow: 0 5px 10px rgba(0,0,0,0.1);
            transition: all 0.3s ease;
        }
        
        .restart-button:hover {
            transform: translateY(-3px) scale(1.05);
            box-shadow: 0 8px 15px rgba(0,0,0,0.2);
        }
        
        .knife-cursor {
            cursor: url('data:image/svg+xml;utf8,<svg xmlns="http://www.w3.org/2000/svg" width="32" height="32" viewBox="0 0 24 24" fill="none" stroke="red" stroke-width="2"><path d="M14.5 8.5L6 17l-3-3 8.5-8.5L14 4l1.5 1.5L17 7l-1.5 1.5z"/><path d="M14.5 8.5L16 10l5.5-5.5-1.5-1.5L14.5 8.5z"/></svg>') 0 24, auto;
        }
        
        .slice-highlight {
            position: absolute;
            pointer-events: none;
            z-index: 5;
            opacity: 0;
            transition: opacity 0.3s ease;
        }
        
        @media (max-width: 850px) {
            .game-container {
                width: 95vw;
                height: 95vw;
                max-width: 800px;
                max-height: 800px;
            }
            
            h1 {
                font-size: 24px;
            }
            
            .instructions {
                font-size: 16px;
                padding: 10px;
            }
            
            .progress-bar {
                width: 95vw;
                max-width: 780px;
            }
            
            .step-circle {
                width: 40px;
                height: 40px;
                font-size: 16px;
            }
            
            .step-label {
                font-size: 12px;
            }
        }
    </style>
</head>
<body>
    <h1>Fraction Fun: Slice the Circle!</h1>
    
    <div class="instructions">
        <p>Drag your finger or mouse across the circle to slice it into equal parts!</p>
        <p>Try to make the fraction shown in the challenge box.</p>
        <p>Hint: Try to slice through the center for the most accurate fractions.</p>
    </div>
    
    <div class="progress-bar" id="progressBar">
        <!-- Progress steps will be added here dynamically -->
    </div>
    
    <div class="game-container">
        <div class="canvas-container knife-cursor">
            <canvas id="gameCanvas" width="800" height="800"></canvas>
        </div>
        
        <div class="challenge-container">
            Challenge: Cut the circle into <span id="fractionText">1/2</span>
        </div>
        
        <div class="fraction-visual" id="fractionVisual">
            <p>This is what <span id="fractionVisualText">1/2</span> looks like:</p>
            <svg id="fractionSvg" viewBox="0 0 100 100">
                <!-- Fraction visualization will be drawn here -->
            </svg>
        </div>
        
        <div class="feedback-container" id="feedback">
            <p>Great job! ðŸŽ‰</p>
            <p>You made <span id="feedbackFraction">1/2</span>!</p>
        </div>
        
        <div class="progress-container">
            Completed: <span id="completedChallenges">0</span>/5
        </div>
        
        <button class="button reset-button" id="resetButton">Reset Circle</button>
        <button class="button next-button" id="nextButton">Next Challenge</button>
        <button class="button hint-button" id="hintButton">Show Hint</button>
    </div>

    <script>
        // Main game variables
        const canvas = document.getElementById('gameCanvas');
        const ctx = canvas.getContext('2d');
        const fractionText = document.getElementById('fractionText');
        const feedbackElement = document.getElementById('feedback');
        const feedbackFraction = document.getElementById('feedbackFraction');
        const completedChallenges = document.getElementById('completedChallenges');
        const resetButton = document.getElementById('resetButton');
        const nextButton = document.getElementById('nextButton');
        const hintButton = document.getElementById('hintButton');
        const fractionVisual = document.getElementById('fractionVisual');
        const fractionVisualText = document.getElementById('fractionVisualText');
        const fractionSvg = document.getElementById('fractionSvg');
        const progressBar = document.getElementById('progressBar');
        
        // Game state
        const gameState = {
            centerX: canvas.width / 2,
            centerY: canvas.height / 2,
            radius: 250,
            pieces: [], // Store all pieces of the circle
            sliceColors: ['#FFD700', '#FF6347', '#32CD32', '#4169E1', '#9370DB', '#FF69B4', '#20B2AA', '#FFA500'],
            isSlicing: false,
            sliceStart: { x: 0, y: 0 },
            sliceEnd: { x: 0, y: 0 },
            selectedPiece: null, // Currently selected piece for slicing
            currentChallenge: 0,
            challenges: [
                { fraction: '1/2', denominator: 2, completed: false },
                { fraction: '1/4', denominator: 4, completed: false },
                { fraction: '1/3', denominator: 3, completed: false },
                { fraction: '1/6', denominator: 6, completed: false },
                { fraction: '1/8', denominator: 8, completed: false }
            ],
            completedCount: 0,
            animationInProgress: false,
            highlightedPiece: null,
            autoProgressTimer: null
        };
        
        // Initialize the game
        function init() {
            // Create initial full circle piece
            resetCircle();
            
            // Set up event listeners
            canvas.addEventListener('mousedown', startSlice);
            canvas.addEventListener('mousemove', handleMouseMove);
            canvas.addEventListener('mouseup', endSlice);
            canvas.addEventListener('touchstart', handleTouchStart);
            canvas.addEventListener('touchmove', handleTouchMove);
            canvas.addEventListener('touchend', handleTouchEnd);
            resetButton.addEventListener('click', resetCircle);
            nextButton.addEventListener('click', moveToNextChallenge);
            hintButton.addEventListener('click', toggleHint);
            
            // Create progress bar
            createProgressBar();
            
            // Set first challenge
            setChallenge(0);
            
            // Create fraction visual
            updateFractionVisual();
            
            // Add intro animation
            showIntroAnimation();
        }
        
        // Create progress bar
        function createProgressBar() {
            progressBar.innerHTML = '';
            
            gameState.challenges.forEach((challenge, index) => {
                const stepElement = document.createElement('div');
                stepElement.className = 'progress-step';
                stepElement.dataset.index = index;
                
                const circleElement = document.createElement('div');
                circleElement.className = 'step-circle';
                if (index === gameState.currentChallenge) {
                    circleElement.classList.add('active');
                }
                if (challenge.completed) {
                    circleElement.classList.add('completed');
                }
                circleElement.textContent = index + 1;
                
                const labelElement = document.createElement('div');
                labelElement.className = 'step-label';
                labelElement.textContent = challenge.fraction;
                
                stepElement.appendChild(circleElement);
                stepElement.appendChild(labelElement);
                
                stepElement.addEventListener('click', () => {
                    setChallenge(index);
                    resetCircle();
                });
                
                progressBar.appendChild(stepElement);
            });
        }
        
        // Update progress bar
        function updateProgressBar() {
            const steps = progressBar.querySelectorAll('.progress-step');
            
            steps.forEach((step, index) => {
                const circle = step.querySelector('.step-circle');
                circle.classList.remove('active', 'completed');
                
                if (index === gameState.currentChallenge) {
                    circle.classList.add('active');
                }
                if (gameState.challenges[index].completed) {
                    circle.classList.add('completed');
                }
            });
        }
        
        // Show intro animation
        function showIntroAnimation() {
            const gameContainer = document.querySelector('.game-container');
            gameContainer.style.opacity = '0';
            gameContainer.style.transform = 'translateY(50px)';
            
            setTimeout(() => {
                gameContainer.style.transition = 'all 1s ease';
                gameContainer.style.opacity = '1';
                gameContainer.style.transform = 'translateY(0)';
            }, 300);
        }
        
        // Draw the circle
        function drawCircle() {
            // Clear canvas
            ctx.clearRect(0, 0, canvas.width, canvas.height);
            
            // Draw center point for reference
            ctx.beginPath();
            ctx.arc(gameState.centerX, gameState.centerY, 5, 0, Math.PI * 2);
            ctx.fillStyle = '#333';
            ctx.fill();
            
            // Draw all pieces
            drawPieces();
            
            // Draw slicing line if in progress
            if (gameState.isSlicing && gameState.selectedPiece !== null) {
                drawSlicingLine();
            }
        }
        
        // Draw all pieces
        function drawPieces() {
            gameState.pieces.forEach((piece, index) => {
                // Draw piece shadow
                ctx.save();
                ctx.beginPath();
                ctx.moveTo(gameState.centerX + piece.offsetX + 3, gameState.centerY + piece.offsetY + 3);
                ctx.arc(
                    gameState.centerX + piece.offsetX + 3, 
                    gameState.centerY + piece.offsetY + 3, 
                    gameState.radius, 
                    piece.startAngle, 
                    piece.endAngle
                );
                ctx.closePath();
                ctx.fillStyle = 'rgba(0, 0, 0, 0.1)';
                ctx.fill();
                ctx.restore();
                
                // Draw piece
                ctx.beginPath();
                ctx.moveTo(gameState.centerX + piece.offsetX, gameState.centerY + piece.offsetY);
                ctx.arc(
                    gameState.centerX + piece.offsetX, 
                    gameState.centerY + piece.offsetY, 
                    gameState.radius, 
                    piece.startAngle, 
                    piece.endAngle
                );
                ctx.closePath();
                
                // Create gradient for piece
                const midAngle = getMidAngle(piece.startAngle, piece.endAngle);
                
                const gradientStartX = gameState.centerX + piece.offsetX - Math.cos(midAngle) * 50;
                const gradientStartY = gameState.centerY + piece.offsetY - Math.sin(midAngle) * 50;
                
                const gradient = ctx.createRadialGradient(
                    gradientStartX, gradientStartY, 0,
                    gameState.centerX + piece.offsetX, gameState.centerY + piece.offsetY, gameState.radius
                );
                
                // Create lighter version of base color
                const baseColor = piece.color;
                const lighterColor = getLighterColor(baseColor);
                
                gradient.addColorStop(0, lighterColor);
                gradient.addColorStop(1, baseColor);
                
                ctx.fillStyle = gradient;
                ctx.fill();
                
                // Draw piece border
                ctx.strokeStyle = '#333';
                ctx.lineWidth = 2;
                ctx.stroke();
                
                // Highlight selected piece
                if (gameState.selectedPiece === index) {
                    ctx.beginPath();
                    ctx.moveTo(gameState.centerX + piece.offsetX, gameState.centerY + piece.offsetY);
                    ctx.arc(
                        gameState.centerX + piece.offsetX, 
                        gameState.centerY + piece.offsetY, 
                        gameState.radius, 
                        piece.startAngle, 
                        piece.endAngle
                    );
                    ctx.closePath();
                    ctx.strokeStyle = 'rgba(255, 255, 255, 0.5)';
                    ctx.lineWidth = 4;
                    ctx.stroke();
                }
                
                // Highlight hovered piece
                if (gameState.highlightedPiece === index && gameState.selectedPiece !== index) {
                    ctx.beginPath();
                    ctx.moveTo(gameState.centerX + piece.offsetX, gameState.centerY + piece.offsetY);
                    ctx.arc(
                        gameState.centerX + piece.offsetX, 
                        gameState.centerY + piece.offsetY, 
                        gameState.radius, 
                        piece.startAngle, 
                        piece.endAngle
                    );
                    ctx.closePath();
                    ctx.strokeStyle = 'rgba(255, 255, 0, 0.5)';
                    ctx.lineWidth = 4;
                    ctx.stroke();
                }
            });
        }
        
        // Get mid angle between two angles, handling the wrap-around case
        function getMidAngle(startAngle, endAngle) {
            let midAngle;
            
            if (endAngle < startAngle) {
                // Handle the case where the piece crosses the 0/2Ï€ boundary
                midAngle = (startAngle + endAngle + 2 * Math.PI) / 2;
                if (midAngle > 2 * Math.PI) {
                    midAngle -= 2 * Math.PI;
                }
            } else {
                midAngle = (startAngle + endAngle) / 2;
            }
            
            return midAngle;
        }
        
        // Get lighter version of a color
        function getLighterColor(hexColor) {
            // Convert hex to RGB
            let r = parseInt(hexColor.substring(1, 3), 16);
            let g = parseInt(hexColor.substring(3, 5), 16);
            let b = parseInt(hexColor.substring(5, 7), 16);
            
            // Lighten
            r = Math.min(255, r + 50);
            g = Math.min(255, g + 50);
            b = Math.min(255, b + 50);
            
            // Convert back to hex
            return `#${r.toString(16).padStart(2, '0')}${g.toString(16).padStart(2, '0')}${b.toString(16).padStart(2, '0')}`;
        }
        
        // Draw the slicing line
        function drawSlicingLine() {
            if (gameState.selectedPiece === null) return;
            
            // Draw main cutting line
            ctx.beginPath();
            ctx.moveTo(gameState.sliceStart.x, gameState.sliceStart.y);
            ctx.lineTo(gameState.sliceEnd.x, gameState.sliceEnd.y);
            
            // Create gradient for line
            const gradient = ctx.createLinearGradient(
                gameState.sliceStart.x, gameState.sliceStart.y,
                gameState.sliceEnd.x, gameState.sliceEnd.y
            );
            gradient.addColorStop(0, '#FF3333');
            gradient.addColorStop(1, '#FF6666');
            
            ctx.strokeStyle = gradient;
            ctx.lineWidth = 3;
            ctx.stroke();
        }
        
        // Set the current challenge
        function setChallenge(index) {
            // Clear any existing auto-progress timer
            if (gameState.autoProgressTimer) {
                clearTimeout(gameState.autoProgressTimer);
                gameState.autoProgressTimer = null;
            }
            
            gameState.currentChallenge = index;
            fractionText.textContent = gameState.challenges[index].fraction;
            fractionVisualText.textContent = gameState.challenges[index].fraction;
            nextButton.style.display = 'none';
            updateFractionVisual();
            updateProgressBar();
            
            // Animate challenge container
            const challengeContainer = document.querySelector('.challenge-container');
            challengeContainer.style.animation = 'none';
            void challengeContainer.offsetWidth; // Trigger reflow
            challengeContainer.style.animation = 'pulse 2s infinite';
        }
        
        // Update fraction visual
        function updateFractionVisual() {
            const denominator = gameState.challenges[gameState.currentChallenge].denominator;
            
            // Clear previous visualization
            fractionSvg.innerHTML = '';
            
            // Create circle with slices
            const circle = document.createElementNS('http://www.w3.org/2000/svg', 'circle');
            circle.setAttribute('cx', '50');
            circle.setAttribute('cy', '50');
            circle.setAttribute('r', '45');
            circle.setAttribute('fill', 'none');
            circle.setAttribute('stroke', '#333');
            circle.setAttribute('stroke-width', '2');
            fractionSvg.appendChild(circle);
            
            // Add slice lines
            for (let i = 0; i < denominator; i++) {
                const angle = (i * 2 * Math.PI) / denominator;
                const x = 50 + 45 * Math.cos(angle);
                const y = 50 + 45 * Math.sin(angle);
                
                const line = document.createElementNS('http://www.w3.org/2000/svg', 'line');
                line.setAttribute('x1', '50');
                line.setAttribute('y1', '50');
                line.setAttribute('x2', x);
                line.setAttribute('y2', y);
                line.setAttribute('stroke', '#333');
                line.setAttribute('stroke-width', '2');
                fractionSvg.appendChild(line);
            }
            
            // Highlight one slice to represent the numerator (always 1 in our case)
            const path = document.createElementNS('http://www.w3.org/2000/svg', 'path');
            const startAngle = 0;
            const endAngle = (2 * Math.PI) / denominator;
            
            const x1 = 50 + 45 * Math.cos(startAngle);
            const y1 = 50 + 45 * Math.sin(startAngle);
            const x2 = 50 + 45 * Math.cos(endAngle);
            const y2 = 50 + 45 * Math.sin(endAngle);
            
            const largeArcFlag = endAngle - startAngle > Math.PI ? 1 : 0;
            
            path.setAttribute('d', `M 50 50 L ${x1} ${y1} A 45 45 0 ${largeArcFlag} 1 ${x2} ${y2} Z`);
            path.setAttribute('fill', '#FFD700');
            path.setAttribute('stroke', '#333');
            path.setAttribute('stroke-width', '1');
            fractionSvg.appendChild(path);
        }
        
        // Toggle hint
        function toggleHint() {
            if (fractionVisual.style.display === 'none' || fractionVisual.style.display === '') {
                fractionVisual.style.display = 'block';
                
                // Animate hint button
                hintButton.style.animation = 'pulse 0.5s';
                setTimeout(() => {
                    hintButton.style.animation = '';
                }, 500);
                
                // Auto-hide hint after 5 seconds
                setTimeout(() => {
                    fractionVisual.style.display = 'none';
                }, 5000);
            } else {
                fractionVisual.style.display = 'none';
            }
        }
        
        // Handle mouse move
        function handleMouseMove(e) {
            if (gameState.isSlicing) {
                updateSlice(e);
            } else {
                // Highlight piece under mouse
                const rect = canvas.getBoundingClientRect();
                const mouseX = e.clientX - rect.left;
                const mouseY = e.clientY - rect.top;
                
                highlightPieceUnderMouse(mouseX, mouseY);
            }
        }
        
        // Highlight piece under mouse
        function highlightPieceUnderMouse(x, y) {
            // Find which piece the mouse is over
            const pieceIndex = findPieceAtPosition(x, y);
            
            if (pieceIndex !== gameState.highlightedPiece) {
                gameState.highlightedPiece = pieceIndex;
                
                // Update cursor style
                if (pieceIndex !== null) {
                    canvas.style.cursor = 'pointer';
                } else {
                    canvas.style.cursor = 'default';
                }
                
                drawCircle();
            }
        }
        
        // Find which piece is at a given position
        function findPieceAtPosition(x, y) {
            // Calculate distance from center
            const distToCenter = Math.sqrt(
                Math.pow(x - gameState.centerX, 2) + 
                Math.pow(y - gameState.centerY, 2)
            );
            
            // If outside the circle, return null
            if (distToCenter > gameState.radius) {
                return null;
            }
            
            // Calculate angle
            let angle = Math.atan2(y - gameState.centerY, x - gameState.centerX);
            if (angle < 0) angle += Math.PI * 2;
            
            // Find which piece contains this angle
            for (let i = 0; i < gameState.pieces.length; i++) {
                const piece = gameState.pieces[i];
                let startAngle = piece.startAngle;
                let endAngle = piece.endAngle;
                
                // Handle pieces that cross the 0/2Ï€ boundary
                if (endAngle < startAngle) {
                    // Check if angle is within this piece
                    if (angle >= startAngle || angle <= endAngle) {
                        return i;
                    }
                } else {
                    // Check if angle is within this piece
                    if (angle >= startAngle && angle <= endAngle) {
                        return i;
                    }
                }
            }
            
            return null;
        }
        
        // Start slicing (mouse event)
        function startSlice(e) {
            if (gameState.animationInProgress) return;
            
            const rect = canvas.getBoundingClientRect();
            const mouseX = e.clientX - rect.left;
            const mouseY = e.clientY - rect.top;
            
            // Find which piece was clicked
            const pieceIndex = findPieceAtPosition(mouseX, mouseY);
            
            if (pieceIndex !== null) {
                gameState.isSlicing = true;
                gameState.selectedPiece = pieceIndex;
                gameState.sliceStart = { x: mouseX, y: mouseY };
                gameState.sliceEnd = { x: mouseX, y: mouseY };
                
                drawCircle();
            }
        }
        
        // Update slicing line (mouse event)
        function updateSlice(e) {
            if (!gameState.isSlicing || gameState.animationInProgress) return;
            
            const rect = canvas.getBoundingClientRect();
            const mouseX = e.clientX - rect.left;
            const mouseY = e.clientY - rect.top;
            
            gameState.sliceEnd = { x: mouseX, y: mouseY };
            
            drawCircle();
        }
        
        // End slicing (mouse event)
        function endSlice(e) {
            if (!gameState.isSlicing || gameState.animationInProgress) return;
            
            const rect = canvas.getBoundingClientRect();
            const mouseX = e.clientX - rect.left;
            const mouseY = e.clientY - rect.top;
            
            gameState.sliceEnd = { x: mouseX, y: mouseY };
            gameState.isSlicing = false;
            
            processSlice();
        }
        
        // Handle touch events
        function handleTouchStart(e) {
            e.preventDefault();
            if (gameState.animationInProgress) return;
            
            const rect = canvas.getBoundingClientRect();
            const touch = e.touches[0];
            const touchX = touch.clientX - rect.left;
            const touchY = touch.clientY - rect.top;
            
            // Find which piece was touched
            const pieceIndex = findPieceAtPosition(touchX, touchY);
            
            if (pieceIndex !== null) {
                gameState.isSlicing = true;
                gameState.selectedPiece = pieceIndex;
                gameState.sliceStart = { x: touchX, y: touchY };
                gameState.sliceEnd = { x: touchX, y: touchY };
                
                drawCircle();
            }
        }
        
        function handleTouchMove(e) {
            e.preventDefault();
            
            if (!gameState.isSlicing || gameState.animationInProgress) return;
            
            const rect = canvas.getBoundingClientRect();
            const touch = e.touches[0];
            const touchX = touch.clientX - rect.left;
            const touchY = touch.clientY - rect.top;
            
            gameState.sliceEnd = { x: touchX, y: touchY };
            
            drawCircle();
        }
        
        function handleTouchEnd(e) {
            e.preventDefault();
            
            if (!gameState.isSlicing || gameState.animationInProgress) return;
            
            gameState.isSlicing = false;
            processSlice();
        }
        
        // Process the slice
        function processSlice() {
            if (gameState.selectedPiece === null) return;
            
            // Calculate if the slice is long enough
            const dx = gameState.sliceEnd.x - gameState.sliceStart.x;
            const dy = gameState.sliceEnd.y - gameState.sliceStart.y;
            const length = Math.sqrt(dx * dx + dy * dy);
            
            // Only process if the slice is long enough (more forgiving)
            if (length < 30) {
                gameState.selectedPiece = null;
                drawCircle();
                return;
            }
            
            // Calculate angle of the slice
            let sliceAngle = Math.atan2(dy, dx);
            if (sliceAngle < 0) sliceAngle += Math.PI * 2;
            
            // Get the selected piece
            const piece = gameState.pieces[gameState.selectedPiece];
            
            // Calculate the two new pieces
            // We need to ensure the angles are properly set to avoid pieces disappearing
            let newStartAngle1 = piece.startAngle;
            let newEndAngle1 = sliceAngle;
            let newStartAngle2 = sliceAngle;
            let newEndAngle2 = piece.endAngle;
            
            // Normalize angles to [0, 2Ï€)
            if (newStartAngle1 < 0) newStartAngle1 += Math.PI * 2;
            if (newEndAngle1 < 0) newEndAngle1 += Math.PI * 2;
            if (newStartAngle2 < 0) newStartAngle2 += Math.PI * 2;
            if (newEndAngle2 < 0) newEndAngle2 += Math.PI * 2;
            
            // Handle the case where the slice angle is outside the piece's angle range
            // This ensures no pieces disappear
            if (piece.endAngle < piece.startAngle) {
                // The piece crosses the 0/2Ï€ boundary
                if (!(sliceAngle >= piece.startAngle || sliceAngle <= piece.endAngle)) {
                    // Slice angle is outside the piece, adjust it to the nearest boundary
                    const distToStart = Math.abs(sliceAngle - piece.startAngle);
                    const distToEnd = Math.abs(sliceAngle - piece.endAngle);
                    const distToEndWrapped = Math.abs(sliceAngle - (piece.endAngle + 2 * Math.PI));
                    
                    if (distToStart < Math.min(distToEnd, distToEndWrapped)) {
                        sliceAngle = piece.startAngle;
                    } else if (distToEnd < distToEndWrapped) {
                        sliceAngle = piece.endAngle;
                    } else {
                        sliceAngle = piece.endAngle;
                    }
                    
                    newEndAngle1 = sliceAngle;
                    newStartAngle2 = sliceAngle;
                }
            } else {
                // Normal piece
                if (sliceAngle < piece.startAngle || sliceAngle > piece.endAngle) {
                    // Slice angle is outside the piece, adjust it to the nearest boundary
                    if (Math.abs(sliceAngle - piece.startAngle) < Math.abs(sliceAngle - piece.endAngle)) {
                        sliceAngle = piece.startAngle;
                    } else {
                        sliceAngle = piece.endAngle;
                    }
                    
                    newEndAngle1 = sliceAngle;
                    newStartAngle2 = sliceAngle;
                }
            }
            
            // Create two new pieces
            const newPiece1 = {
                startAngle: newStartAngle1,
                endAngle: newEndAngle1,
                offsetX: 0,
                offsetY: 0,
                color: piece.color
            };
            
            const newPiece2 = {
                startAngle: newStartAngle2,
                endAngle: newEndAngle2,
                offsetX: 0,
                offsetY: 0,
                color: gameState.sliceColors[(gameState.pieces.length + 1) % gameState.sliceColors.length]
            };
            
            // Check if the pieces are valid (have a non-zero angle)
            let angle1 = newEndAngle1 - newStartAngle1;
            if (angle1 < 0) angle1 += Math.PI * 2;
            
            let angle2 = newEndAngle2 - newStartAngle2;
            if (angle2 < 0) angle2 += Math.PI * 2;
            
            // Only proceed if both pieces have a valid angle
            if (angle1 > 0.01 && angle2 > 0.01) {
                // Remove the original piece and add the new ones
                gameState.pieces.splice(gameState.selectedPiece, 1, newPiece1, newPiece2);
                
                // Create sparkles at intersection points
                createSparklesAtIntersections(sliceAngle);
                
                // Animate the pieces moving apart
                animatePieces();
            } else {
                // If the slice would create invalid pieces, just reset the selection
                gameState.selectedPiece = null;
                drawCircle();
            }
        }
        
        // Create sparkles at intersection points
        function createSparklesAtIntersections(angle) {
            // Calculate intersection points with the circle
            const x1 = gameState.centerX + Math.cos(angle) * gameState.radius;
            const y1 = gameState.centerY + Math.sin(angle) * gameState.radius;
            const x2 = gameState.centerX + Math.cos(angle + Math.PI) * gameState.radius;
            const y2 = gameState.centerY + Math.sin(angle + Math.PI) * gameState.radius;
            
            // Create sparkles at these points
            createSparkles(x1, y1);
            createSparkles(x2, y2);
        }
        
        // Create sparkles at position
        function createSparkles(x, y) {
            // Create sparkle elements
            for (let i = 0; i < 15; i++) {
                const sparkle = document.createElement('div');
                sparkle.className = 'confetti';
                
                // Random color
                const colors = ['#ff0000', '#00ff00', '#0000ff', '#ffff00', '#ff00ff', '#00ffff'];
                sparkle.style.backgroundColor = colors[Math.floor(Math.random() * colors.length)];
                
                // Position at the sparkle point
                sparkle.style.left = x + 'px';
                sparkle.style.top = y + 'px';
                
                // Random size
                const size = Math.random() * 6 + 2;
                sparkle.style.width = size + 'px';
                sparkle.style.height = size + 'px';
                
                document.querySelector('.game-container').appendChild(sparkle);
                
                // Animate sparkle
                const angle = Math.random() * Math.PI * 2;
                const distance = Math.random() * 50 + 20;
                const duration = Math.random() * 1 + 0.5;
                
                sparkle.style.transition = `all ${duration}s ease-out`;
                
                setTimeout(() => {
                    sparkle.style.transform = `translate(${Math.cos(angle) * distance}px, ${Math.sin(angle) * distance}px)`;
                    sparkle.style.opacity = '0';
                }, 10);
                
                setTimeout(() => {
                    sparkle.remove();
                }, duration * 1000);
            }
        }
        
        // Animate pieces moving apart
        function animatePieces() {
            gameState.animationInProgress = true;
            
            // Animation parameters
            const animationDuration = 500; // ms
            const maxOffset = 10; // pixels
            const startTime = Date.now();
            
            function animationStep() {
                const elapsed = Date.now() - startTime;
                const progress = Math.min(elapsed / animationDuration, 1);
                
                // Ease in-out function
                const easeProgress = progress < 0.5 
                    ? 2 * progress * progress 
                    : 1 - Math.pow(-2 * progress + 2, 2) / 2;
                
                const currentOffset = maxOffset * easeProgress;
                
                // Update piece positions
                // We'll move each piece outward from the center
                gameState.pieces.forEach(piece => {
                    const midAngle = getMidAngle(piece.startAngle, piece.endAngle);
                    
                    piece.offsetX = Math.cos(midAngle) * currentOffset;
                    piece.offsetY = Math.sin(midAngle) * currentOffset;
                });
                
                // Redraw
                drawCircle();
                
                if (progress < 1) {
                    requestAnimationFrame(animationStep);
                } else {
                    gameState.animationInProgress = false;
                    gameState.selectedPiece = null;
                    checkPieceCount();
                }
            }
            
            animationStep();
        }
        
        // Check if we've created the right number of pieces for the challenge
        function checkPieceCount() {
            // For simplicity, we'll just check if the number of pieces matches the denominator
            const targetPieces = gameState.challenges[gameState.currentChallenge].denominator;
            
            if (gameState.pieces.length === targetPieces) {
                // Check if the pieces are roughly equal in size (more forgiving)
                const equalPieces = checkEqualPieces();
                
                if (equalPieces) {
                    // Challenge completed!
                    setTimeout(() => {
                        if (!gameState.challenges[gameState.currentChallenge].completed) {
                            gameState.challenges[gameState.currentChallenge].completed = true;
                            gameState.completedCount++;
                            completedChallenges.textContent = gameState.completedCount;
                            updateProgressBar();
                            
                            // Show confetti
                            createConfetti();
                            
                            // Show feedback
                            feedbackFraction.textContent = gameState.challenges[gameState.currentChallenge].fraction;
                            feedbackElement.style.display = 'block';
                            
                            // Auto move to next challenge after 2 seconds
                            if (gameState.autoProgressTimer) {
                                clearTimeout(gameState.autoProgressTimer);
                            }
                            
                            gameState.autoProgressTimer = setTimeout(() => {
                                // Find next uncompleted challenge
                                if (gameState.completedCount < gameState.challenges.length) {
                                    moveToNextChallenge();
                                } else {
                                    // All challenges completed!
                                    showAllCompletedMessage();
                                }
                                gameState.autoProgressTimer = null;
                            }, 2000);
                        }
                    }, 500);
                }
            }
        }
        
        // Check if pieces are roughly equal in size (more forgiving)
        function checkEqualPieces() {
            // Calculate the area of each piece
            const areas = gameState.pieces.map(piece => {
                let angle = piece.endAngle - piece.startAngle;
                if (angle < 0) angle += Math.PI * 2;
                return angle;
            });
            
            // Calculate the average area
            const avgArea = areas.reduce((sum, area) => sum + area, 0) / areas.length;
            
            // Check if all areas are within 30% of the average (more forgiving)
            const tolerance = 0.3;
            
            for (let area of areas) {
                if (Math.abs(area - avgArea) / avgArea > tolerance) {
                    return false;
                }
            }
            
            return true;
        }
        
        // Show temporary message
        function showTemporaryMessage(message) {
            const tempMessage = document.createElement('div');
            tempMessage.className = 'temp-message';
            tempMessage.textContent = message;
            
            document.querySelector('.game-container').appendChild(tempMessage);
            
            setTimeout(() => {
                tempMessage.remove();
            }, 1500);
        }
        
        // Create confetti effect
        function createConfetti() {
            const confettiCount = 100;
            const colors = ['#ff0000', '#00ff00', '#0000ff', '#ffff00', '#ff00ff', '#00ffff'];
            
            for (let i = 0; i < confettiCount; i++) {
                const confetti = document.createElement('div');
                confetti.className = 'confetti';
                confetti.style.backgroundColor = colors[Math.floor(Math.random() * colors.length)];
                confetti.style.left = Math.random() * canvas.width + 'px';
                confetti.style.top = -10 + 'px';
                confetti.style.width = Math.random() * 10 + 5 + 'px';
                confetti.style.height = Math.random() * 10 + 5 + 'px';
                confetti.style.opacity = Math.random() + 0.5;
                confetti.style.borderRadius = Math.random() > 0.5 ? '50%' : '0';
                confetti.style.transform = `rotate(${Math.random() * 360}deg)`;
                
                document.querySelector('.game-container').appendChild(confetti);
                
                // Animate confetti
                const animationDuration = Math.random() * 3 + 2;
                const xPos = Math.random() * canvas.width;
                
                confetti.style.transition = `all ${animationDuration}s ease-out`;
                
                setTimeout(() => {
                    confetti.style.transform = `translate(${(Math.random() - 0.5) * 200}px, ${canvas.height + 20}px) rotate(${Math.random() * 360}deg)`;
                    confetti.style.opacity = '0';
                }, 10);
                
                setTimeout(() => {
                    confetti.remove();
                }, animationDuration * 1000);
            }
        }
        
        // Move to next challenge
        function moveToNextChallenge() {
            // Hide feedback
            feedbackElement.style.display = 'none';
            nextButton.style.display = 'none';
            
            // Find next uncompleted challenge
            if (gameState.completedCount < gameState.challenges.length) {
                let nextChallenge = (gameState.currentChallenge + 1) % gameState.challenges.length;
                while (gameState.challenges[nextChallenge].completed && gameState.completedCount < gameState.challenges.length) {
                    nextChallenge = (nextChallenge + 1) % gameState.challenges.length;
                }
                
                setChallenge(nextChallenge);
                resetCircle();
            } else {
                // All challenges completed!
                showAllCompletedMessage();
            }
        }
        
        // Show message when all challenges are completed
        function showAllCompletedMessage() {
            const allCompletedMessage = document.createElement('div');
            allCompletedMessage.className = 'completion-message';
            allCompletedMessage.innerHTML = `
                <h2>Congratulations! ðŸŽ‰</h2>
                <p>You've completed all fraction challenges!</p>
                <p>You're a fraction master!</p>
            `;
            
            document.querySelector('.game-container').appendChild(allCompletedMessage);
            
            // Create lots of confetti
            createConfetti();
            setTimeout(createConfetti, 500);
            setTimeout(createConfetti, 1000);
            
            // Add button to restart
            const restartButton = document.createElement('button');
            restartButton.className = 'restart-button';
            restartButton.textContent = 'Play Again';
            
            restartButton.addEventListener('click', () => {
                allCompletedMessage.remove();
                resetGame();
            });
            
            allCompletedMessage.appendChild(restartButton);
        }
        
        // Reset the circle
        function resetCircle() {
            // Create initial full circle piece
            gameState.pieces = [{
                startAngle: 0,
                endAngle: Math.PI * 2,
                offsetX: 0,
                offsetY: 0,
                color: gameState.sliceColors[0]
            }];
            
            gameState.selectedPiece = null;
            gameState.isSlicing = false;
            gameState.highlightedPiece = null;
            
            // Clear any auto-progress timer
            if (gameState.autoProgressTimer) {
                clearTimeout(gameState.autoProgressTimer);
                gameState.autoProgressTimer = null;
            }
            
            // Draw the circle
            drawCircle();
            
            // Hide fraction visual if shown
            fractionVisual.style.display = 'none';
            
            // Animate reset button
            resetButton.style.animation = 'pulse 0.5s';
            setTimeout(() => {
                resetButton.style.animation = '';
            }, 500);
        }
        
        // Reset the entire game
        function resetGame() {
            // Create initial full circle piece
            gameState.pieces = [{
                startAngle: 0,
                endAngle: Math.PI * 2,
                offsetX: 0,
                offsetY: 0,
                color: gameState.sliceColors[0]
            }];
            
            gameState.selectedPiece = null;
            gameState.isSlicing = false;
            gameState.highlightedPiece = null;
            gameState.currentChallenge = 0;
            gameState.completedCount = 0;
            
            // Clear any auto-progress timer
            if (gameState.autoProgressTimer) {
                clearTimeout(gameState.autoProgressTimer);
                gameState.autoProgressTimer = null;
            }
            
            // Reset challenges
            gameState.challenges.forEach(challenge => {
                challenge.completed = false;
            });
            
            completedChallenges.textContent = '0';
            
            // Update progress bar
            updateProgressBar();
            
            // Set first challenge
            setChallenge(0);
            
            // Draw initial circle
            drawCircle();
            
            // Hide fraction visual if shown
            fractionVisual.style.display = 'none';
        }
        
        // Initialize the game
        init();
    </script>
</body>
</html>


==============

Gen cÃ¡i game HAY HÆ N GAME NÃ€Y 100 Láº¦N 
```
